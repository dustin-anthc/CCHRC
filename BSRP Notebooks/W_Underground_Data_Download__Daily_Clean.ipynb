```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matpotlib inline
```

```python
def download_daily_weather(weather_station, year, month, day):
    
    url_template = "http://www.wunderground.com/history/airport/{WEATHERSTATION}/{YEAR}/{MONTH}/{DAY}/MonthlyHistory.html?format=1"
    url = url_template.format(WEATHERSTATION=weather_station, DAY=day, MONTH=month, YEAR=year)
    daily_weather = pd.read_csv(url, header=1)
    column_head = daily_weather.columns.values.tolist()

    if column_head[0] == 'AKST':
        daily_weather = daily_weather.rename(columns = {'AKST' : 'AKDT'})
    
    daily_weather = daily_weather.dropna(axis=0, how='all')
    daily_weather = daily_weather.drop([' CloudCover', ' Events', ' Max Gust SpeedMPH', ' Max Sea Level PressureIn',
                                        ' Max VisibilityMiles', ' Max Wind SpeedMPH', ' Mean Humidity', ' Mean Sea Level PressureIn',
                                        ' Mean VisibilityMiles',' Mean Wind SpeedMPH',' Min Humidity',' Min Sea Level PressureIn',
                                        ' Min VisibilityMiles',' WindDirDegrees<br />','Max Dew PointF','Max Humidity',
                                        'MeanDew PointF','Min DewpointF','PrecipitationIn'], axis=1)
    
    return daily_weather
```

```python
# Need to set year and weather station manually.
weather_station = 'PAWG'
year = 2014
yearly_data = [download_daily_weather(weather_station, year, i, 1) for i in range (1,12)]
#yearly_data = [download_daily_weather(year, i, 13) for i in range (1, 12)]
yearly_weather = pd.concat(yearly_data)
yearly_weather
```

```python
# Make sure to put in a new file name
filename = str(year) + '_' + weather_station + '_Annual_Daily_Weather.xlsx'
yearly_weather.to_excel(filename)
```
