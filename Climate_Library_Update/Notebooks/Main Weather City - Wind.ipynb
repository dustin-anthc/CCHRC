```python
subset = tmy_ids[['ids', 'noaa_index']]
tmy_tuple = [tuple(x) for x in subset.values]
tmy_tuple
```

```python
# tmy_ids = pd.DataFrame(tmy_ids, columns=['id','station_name','tmy_class'])

tmy_first_order = pd.DataFrame()

for ids, noaa_index, in tmy_tuple:
 
    # Read the TMY3 file into a Pandas DataFrame direct from the NREL site
    tmy = pd.read_csv('http://rredc.nrel.gov/solar/old_data/nsrdb/1991-2005/data/tmy3/%sTYA.CSV' % ids, skiprows=1)
    # Create column with noaa station index
    tmy['STATION_NAME'] = noaa_index
    tmy_first_order = tmy_first_order.append(tmy)

tmy_first_order.head()
```

```python
# Changes date to pandas DateTime format
tmy_first_order['Date (MM/DD/YYYY)'] = to_datetime(tmy_first_order['Date (MM/DD/YYYY)'])
# Creates month column
tmy_first_order['MONTH'] = pd.DatetimeIndex(tmy_first_order['Date (MM/DD/YYYY)']).month
tmy_first_order.head()
```

```python
# Converts hourly wind speeds from meters per second to miles per hour in a new column
tmy_first_order['wind_speed_mph'] = tmy_first_order['Wspd (m/s)'].map(lambda x: float(x) * 60 * 60 / 1609.34)
```

```python
# Calculate average wind speed by month for each station
tmy_monthly_wind = tmy_first_order.groupby(['STATION_NAME', 'MONTH']).mean()
tmy_monthly_wind = tmy_monthly_wind[['wind_speed_mph']]
tmy_monthly_wind.head()
```

```python
#tmy_monthly_wind.to_excel('TMY_Wind_Speed_Averages.xlsx', sheet_name='TMY3')
```

```python
# Prepare datasets to be combined
tmy_right = tmy_monthly_wind
noaa_left = NOAA52
tmy_right.reset_index(inplace=True)
tmy_right = tmy_right.rename(columns = {'wind_speed_mph' : 'tmy_wind_speed'})
tmy_right.head()
```

```python
noaa_left = noaa_left.rename(columns = {'WINDSPEED_MPH' : 'noaa_wind_speed'})
noaa_left.head()
```

```python
# Combine the NOAA nad TMY datasets
wind_joined = pd.merge(noaa_left, tmy_right, left_on=['station_name'], 
                        right_on=['STATION_NAME'], how='left')
#wind_compare = wind_compare[['STATION_NAME', 'MONTH', 'tmy_wind_speed', 'noaa_wind_speed']]
wind_joined.head()
```

```python
wind_joined['MONTH'] = np.where(wind_joined['MONTH_x'] == wind_joined['MONTH_x'], wind_joined['MONTH_x'], wind_joined['MONTH_y'])
wind_joined.head()
```

```python
wind_joined['data_source'] = np.where(wind_joined['tmy_wind_speed'] == wind_joined['tmy_wind_speed'], 'tmy3', 'no_data')
wind_joined['data_source'] = np.where(wind_joined['noaa_wind_speed'] == wind_joined['noaa_wind_speed'], 'noaa', wind_joined['data_source'])

wind_joined['wind_speed_mph'] = np.where(wind_joined['noaa_wind_speed'] == wind_joined['noaa_wind_speed'], wind_joined['noaa_wind_speed'], 'no_data')
wind_joined['wind_speed_mph'] = np.where(wind_joined['tmy_wind_speed'] == wind_joined['tmy_wind_speed'], wind_joined['tmy_wind_speed'], wind_joined['wind_speed_mph'])
wind_joined.head()
```

```python
scaled_stations['ref_data_source'] = scaled_stations['data_source'].str.split('(').str.get(1)
scaled_stations['ref_data_source'] = scaled_stations['ref_data_source'].str.split(')').str.get(0)
scaled_stations = scaled_stations.drop('data_source', 1)
scaled_stations.head()
```

```python
ref_wind = wind_data[wind_data['station_name'].isin(scaled_stations['Reference_STATION_NAME'])]
ref_wind.head()
```

```python
full_wind = scaled_stations.merge(ref_wind, left_on='Reference_STATION_NAME', right_on='station_name')
full_wind.head()
```

```python
full_wind['wind_speed_mph_float'] = full_wind['wind_speed_mph'].map(lambda x: float(x))
full_wind['adjusted_monthly_wind'] = full_wind['Wind_Multiplier'] * full_wind['wind_speed_mph_float']
full_wind.head()
```

```python
full_wind_df = full_wind[['STATION_NAME', 'MONTH', 'adjusted_monthly_wind']]
full_wind_df.head()
```

```python
all_wind = wind_data.merge(full_wind_df, left_on='station_name', right_on='STATION_NAME', how='left')
all_wind.head()
```

```python
all_wind['MONTH'] = np.where(all_wind['MONTH_x'] == all_wind['MONTH_x'], all_wind['MONTH_x'], 'no_data')
all_wind['MONTH'] = np.where(all_wind['MONTH_y'] == all_wind['MONTH_y'], all_wind['MONTH_y'], all_wind['MONTH'])
all_wind['Wind'] = np.where(all_wind['wind_speed_mph'] == all_wind['wind_speed_mph'], all_wind['wind_speed_mph'], 'no_data')
all_wind['Wind'] = np.where(all_wind['adjusted_monthly_wind'] == all_wind['adjusted_monthly_wind'], all_wind['adjusted_monthly_wind'], all_wind['Wind'])
all_wind.head()
```
