## This notebook combines the data from the Main_Weather_Item spreadsheet with TMY3 solar data with the multiplied solar data supplied by Ian.


```python
monthly_weather = pd.read_excel(r"C:\Users\dustin\Desktop\Monthly_Weather_Items_with_IDs.xlsx")
monthly_weather.head()
```

```python
ian_solar = pd.read_excel(r"C:\Users\dustin\Desktop\Solar_Station_Calculated_Values.xlsx")
ian_solar.head()
```

```python
combined_solar = pd.merge(monthly_weather, ian_solar, how='left', left_on=['station_name', 'MONTH'], right_on=['STATION_NAME', 'Month'], suffixes=('', '_ian'))
combined_solar.head()
```

```python
list(combined_solar.columns.values)
```

```python
tmy_column = ['South_Vert_Solar',
 'East_Vert_Solar',
 'West_Vert_Solar',
 'North_Vert_Solar',
 'HorizSol']

ian_column = ['South_Vert_Solar_ian',
              'East_Vert_Solar_ian',
              'West_Vert_Solar_ian',
              'North_Vert_Solar_ian', 
              'HorizSol_ian']

for i in range(0, len(tmy_columns)):
    combined_solar[tmy_column[i]] = np.where(combined_solar[tmy_column[i]] == combined_solar[tmy_column[i]], 
                                      combined_solar[tmy_column[i]], combined_solar[ian_column[i]])
combined_solar.describe()
```

```python
main_weather_city = combined_solar[['Main_Weather_ID','station_name','MONTH','DBTempAvg','Relative  Humidity','Wind',
                                    'South_Vert_Solar','East_Vert_Solar','West_Vert_Solar','North_Vert_Solar','HorizSol',
                                    'bin data']]
main_weather_city.head()
```

```python
columns = ['Relative  Humidity','Wind']
solar_columns = ['South_Vert_Solar','East_Vert_Solar','West_Vert_Solar','North_Vert_Solar','HorizSol']
for column in columns: 
    main_weather_city[column] = np.round(main_weather_city[column], 1)
for solar_column in solar_columns:
    main_weather_city[solar_column] = np.round(main_weather_city[solar_column], 0)
    
main_weather_city.tail()
```

```python
main_weather_city.to_csv("final_main_weather_city_fixed_solar.csv")
```
