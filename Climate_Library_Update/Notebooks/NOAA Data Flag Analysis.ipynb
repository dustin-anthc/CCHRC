```python
df = df.rename(columns={
        'Completeness Flag.31':'DataQuality'
    })
df.head()
```

```python
good_data_mask = (df['DataQuality'] == 'C') | (df['DataQuality'] == 'S')
good_data = df[good_data_mask]
good_data_num = len(good_data.index)
all_data_num = len(df.index)
percent = float(good_data_num) / float(all_data_num) * 100
print "%d of %d, or approximately %f percent, of the monthly average temperature numbers are considered either complete or standard data." % (good_data_num, all_data_num, percent)
```

```python
good_data_counts = good_data.groupby(['STATION_NAME']).count()
good_data_counts
```

```python
fulldatamask = (good_data_counts['STATION'] == 12)
full_data_stations = good_data_counts[fulldatamask]
full_data_stations
```

```python
communities = df.groupby(['STATION_NAME']).count()
communities
```

```python
full_communities = len(full_data_stations.index)
all_communities = len(communities.index)

print "%d of the %d stations with NOAA 30 year climate normals data have complete or standard data for each month." % (full_communities, all_communities)
```
