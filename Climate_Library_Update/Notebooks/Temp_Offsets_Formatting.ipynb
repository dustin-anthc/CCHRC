```python
subset = offsets[['City_Name', 'Temp_1',
 'Temp_2',
 'Temp_3',
 'Temp_4',
 'Temp_5',
 'Temp_6',
 'Temp_7',
 'Temp_8',
 'Temp_9',
 'Temp_10',
 'Temp_11',
 'Temp_12']]

solar_subset = offsets[['City_Name', 'Solar_1',
 'Solar_2',
 'Solar_3',
 'Solar_4',
 'Solar_5',
 'Solar_6',
 'Solar_7',
 'Solar_8',
 'Solar_9',
 'Solar_10',
 'Solar_11',
 'Solar_12']]
```

```python
subset.head()
```

```python
results = pd.DataFrame()
monthly_offsets = pd.DataFrame()
for x in range(1,13):
    results['City_Name'] = subset[['City_Name']]
    results['Temp_offset'] = subset[subset.columns[x]]
    results['month'] = x
    monthly_offsets = monthly_offsets.append(results)
    
    
```

```python
sorted_offsets = monthly_offsets.sort(columns=['City_Name', 'month'])
sorted_offsets
```

```python
results = pd.DataFrame()
solar_offsets = pd.DataFrame()
for x in range(1,13):
    results['City_Name'] = solar_subset[['City_Name']]
    results['solar_offset'] = solar_subset[solar_subset.columns[x]]
    results['month'] = x
    solar_offsets = solar_offsets.append(results)
    
sorted_solar = solar_offsets.sort(columns=['City_Name', 'month'])
sorted_solar
```

```python
final_offsets = pd.merge(sorted_offsets, sorted_solar, on=['City_Name', 'month'], how='left')
final_offsets.head()
```

```python
all_offsets = pd.merge(final_offsets, offsets, on='City_Name', how='left')
all_offsets
```
