```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
###################### Matplotlib #######################
# import matplotlib pyplot commands
from matplotlib.pyplot import *

import matplotlib
matplotlib.style.use('ggplot')

# Show Plots in the Notebook
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\dustin\Desktop\commercial_outcomes_multiple_regression.csv", header=1)
df.head()
```

```python
df.dtypes
```

```python
clean_df = df[['Com_AkWarmID','City', 'Building Name','ANCSA Region', 'Research Usage Type', 'Age', 'HDD', ' sqft ',
             'Thermal EUI/HDD','DATASET', 'ANNUAL MMBTU']]
clean_df.head()
```

```python
clean_df = clean_df.rename(columns = {'Thermal EUI/HDD':'ThermalEUI', 
                                      ' sqft ':'sqft',
                                      'ANNUAL MMBTU':'ANN_MMBTU',
                                      'Research Usage Type':'Usage_Type'
                                     })
clean_df.head()
```

```python
audit_anthc_df = clean_df.query("DATASET == 'AUDIT' or DATASET == 'ANTHC'")
audit_anthc_df
```

```python
audit_anthc_df.dtypes
```

```python
audit_anthc_df['sqft'] = audit_anthc_df['sqft'].str.replace(',','')
audit_anthc_df['sqft'].unique()
```

```python
audit_anthc_df[['Age', 'HDD', 'sqft']] = audit_anthc_df[['Age', 'HDD', 'sqft']].astype(float)
audit_anthc_df
```

```python
audit_anthc_df.dtypes
```

```python
est = smf.ols(formula='ThermalEUI ~ Age + HDD + sqft', data=audit_anthc_df).fit()
est.summary()
```

```python
schools = audit_anthc_df.query("Usage_Type == 'Education - K - 12'")
schools.head()
```

```python
schools_est = smf.ols(formula='ANN_MMBTU ~ Age + HDD + sqft', data=schools).fit()
schools_est.summary()
```
