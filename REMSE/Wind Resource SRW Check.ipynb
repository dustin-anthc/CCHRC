```python
cleaned_df = df.drop_duplicates(subset=('LocationID', 'Address', 'City', 'FullName', 'RatingType' ,'RatingPoints'))
cleaned_df
```

```python
dropped = df.join(cleaned_df, how='left', rsuffix='.cleaned')
dropped
```

```python
# Add in mask that checks to see if "cleaned" columns are NAs, so new dataset only includes dropped values
mask1 = dropped['LocationID.cleaned'].map(lambda x: np.isnan(x))
dropped = dropped[mask1]
dropped
```
